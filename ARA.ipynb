```python
import scipy.io
import pandas as pd
data = pd.read_csv('ARA.csv')
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = data.tdigitized;
f = 2 * np.exp(-t**2 / 4) * np.sin(2 * np.pi * t)
plt.plot(data.tdigitized, data.digitized, label='Hardware')
plt.plot(t, f, '--', label='Simulation')
plt.legend()
#plt.xlabel('Time')
#plt.ylabel('Amplitude')
#plt.title('Comparison of Hardware and Simulation')
#plt.grid(True)
plt.show()
```
